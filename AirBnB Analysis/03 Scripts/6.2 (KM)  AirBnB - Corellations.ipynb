# 6. Create a categorical plot and interpret the results.

```python
sns.histplot(df['guest_satisfaction_overall'], bins = 20, kde = True)
```

```python
df.loc[df['guest_satisfaction_overall'] < 60, 'satisfaction'] = 'Unsatisfied'
```

```python
df.loc[(df['guest_satisfaction_overall'] >= 60) & (df['guest_satisfaction_overall'] < 90), 'satisfaction'] = 'Satisfied'
```

```python
df.loc[df['guest_satisfaction_overall'] >= 90, 'satisfaction'] = 'Very Satisfied'
```
